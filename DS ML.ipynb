```python
import matplotlib 
import random
from matplotlib import pyplot as plt
import numpy as np

def f(u):
    f= 10 - (140*u) +  400*(u**2) - 250*(u**3)
    return f

M, N = [],[]
for i in range(15):
    k=random.random()
    M.append(k)
    N.append(f(k))
for i in range(70):
    k=random.random()
    l=random.randrange(-15,15)
    M.append(k)
    N.append(f(k)+l)
plt.scatter (M,N)
S,T=[],[]
```

```python
t = np.linspace(0,1,300)
p=np.poly1d(np.polyfit(M,N,1))
plt.plot(t, f(t), color='red')
plt.scatter(M, N)
plt.plot(t, p(t))
p(t)
cf = 0
for i in t:
    cf +=(p(i)-f(i))**2
S.append(1)
T.append(np.sqrt(cf))
print (np.sqrt(cf))
```
