```python
from numpy import genfromtxt
from sklearn. model_selection import train_test_split
import matplotlib .pyplot as plt
from mpl_toolkits .mplot3d import Axes3D
import numpy as np
import pandas as pd
```

```python
pa = pd.read_csv("pricerunner_aggregate.csv")
pa=pa[0:10000]
train,test = train_test_split(pa,test_size= 0.3)
```

```python
X_train = train[['ProductID',' MerchantID',' ClusterID',' CategoryID']]
y_train = train.CategoryLabel
X_test = test[['ProductID',' MerchantID',' ClusterID',' CategoryID']]
y_test = test.CategoryLabel
```

```python
from sklearn.metrics import accuracy_score
import sklearn. discriminant_analysis as DA
from sklearn. naive_bayes import GaussianNB
from sklearn. neighbors import KNeighborsClassifier
from sklearn. linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
names = ["Logit","NBayes", "LDA", "QDA", "KNN", "SVM"]
classifiers = [ LogisticRegression (C=1e4, max_iter=1000000),
               GaussianNB (),
                DA. LinearDiscriminantAnalysis (),
                DA. QuadraticDiscriminantAnalysis (),
                KNeighborsClassifier ( n_neighbors =5),
                SVC()]

print('Name Accuracy\n'+14*'-')

for name , clf in zip(names , classifiers ):
    clf.fit(X_train , y_train)
    y_pred = clf.predict(X_test)
    print('{:6} {:3.3f}'.format(name , accuracy_score (y_test ,y_pred)))
```
