<h3>$Ans \; B3 \; (d)$

```python
L=[]
for i in k:
    if (i<0):
        L.append(0)
    else:
        L.append(i)
L1=[]
for i in k:
        L1.append(np.tanh(i))
```

```python
plt.plot(L,'--',color='magenta',label='ReLU')
plt.plot(L1,color='green',label='tanh')
plt.xlabel('x')
plt.ylabel('ReLU/tanh')
plt.legend()
plt.show()
```
